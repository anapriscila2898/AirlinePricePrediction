```python
#fitting the model using RandomForestRegressor along with FeatureImportance
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
X = df1.drop('Price',axis=1)
y = df1.Price.values
```

```python
X.head()
```

```python
X.shape
```

```python
y
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
```

```python
rf = RandomForestRegressor(n_estimators=100)
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
plt.barh(X.columns, rf.feature_importances_)
```

```python
sorted_idx = rf.feature_importances_.argsort()
plt.barh(X.columns[sorted_idx], rf.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```

```python
y_predict=rf.predict(X_test)
```

```python
metrics.r2_score(y_test,y_predict)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'n_estimators':range(10,100,10),'criterion':('mse','mae'),'max_features':('auto','sqrt','log2')}
```

```python
gs=GridSearchCV(rf,parameters,n_jobs=15)
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
from sklearn.metrics import mean_absolute_error
y_prediction=gs.best_estimator_.predict(X_test)
mean_absolute_error(y_test,y_prediction)
```

```python
metrics.r2_score(y_test,y_prediction)
```
